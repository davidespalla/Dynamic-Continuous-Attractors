# Dynamics visualization
Plots of three different snaphsot of the population dynamics, with a single stored map, for one dimensional, two dimensional and three dimensional DCAs.

## 1D

```python
pfc=load("1D/pfc.npy")
Vs=load("1D/Vdynamics.npy")
```

```python
t=49
Dt=2

color=cm.Blues(np.linspace(0.5,1,3))

figure(figsize=(6,6))
scatter(pfc[0],Vs[t-Dt],label=r"$t_{1}$",c=color[0],s=10)
scatter(pfc[0],Vs[t],label=r"$t_{2}$",c=color[1],s=10)
scatter(pfc[0],Vs[t+Dt],label=r"$t_{3}$",c=color[2],s=10)
xlabel("x (a.u.)", fontsize=20)
ylabel("V (a.u.)", fontsize=20)
#savefig("1D.svg")
#legend()
```

## 2D

```python
pfc=load("2D/pfc.npy")
Vs=load("2D/Vdynamics.npy")
```

```python
t=60
Dt=4

color=cm.Blues(np.linspace(0.5,1,3))

x = np.arange(0, 10, 10.0/30.0)
y = np.arange(0, 10, 10.0/30.0)
X, Y = np.meshgrid(x, y)
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.set_xlabel("x (a.u.)",fontsize=20)
ylabel("y (a.u.)",fontsize=20)
ax.set_zlabel("V (a.u.)",fontsize=20)
#ax.grid("off")
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False


Z = Vs[t-Dt].reshape((30,30))
ax.plot_surface(X, Y, Z,alpha=0.6,color=color[0],label=r"$t_{1}$")

Z = Vs[t].reshape((30,30))
ax.plot_surface(X, Y, Z,alpha=0.6,color=color[1],label=r"$t_{2}$")

Z = Vs[t+Dt].reshape((30,30))
ax.plot_surface(X, Y, Z,alpha=0.6,color=color[2],label=r"$t_{3}$")

#savefig("2D.svg")
```

## 3D

```python
pfc=load("3D/pfc.npy")
Vs=load("3D/Vdynamics.npy")
```

```python
t=51
Dt=3

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')

ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False


ax.set_xlim3d(0,5)
ax.set_ylim3d(0,5)
ax.set_zlim3d(0,5)
    
    
ax.set_xlabel('x (a.u.)',fontsize=20)
ax.set_ylabel('y (a.u.)',fontsize=20)
ax.set_zlabel('z (a.u.)',fontsize=20)
    
V=[]
x=[]
y=[]
z=[]
for j in range(len(Vs[t-Dt])):
    if Vs[t-Dt][j]>0:
        V.append(Vs[t-Dt][j])
        x.append(pfc[0].T[0][j])
        y.append(pfc[0].T[1][j])
        z.append(pfc[0].T[2][j])
    
ax.scatter(x,y,z,c=V,cmap=cm.jet,s=10)

V=[]
x=[]
y=[]
z=[]
for j in range(len(Vs[t])):
    if Vs[t][j]>0:
        V.append(Vs[t][j])
        x.append(pfc[0].T[0][j])
        y.append(pfc[0].T[1][j])
        z.append(pfc[0].T[2][j])
    
ax.scatter(x,y,z,c=V,cmap=cm.jet,s=10)

V=[]
x=[]
y=[]
z=[]
for j in range(len(Vs[t+Dt])):
    if Vs[t+Dt][j]>0:
        V.append(Vs[t+Dt][j])
        x.append(pfc[0].T[0][j])
        y.append(pfc[0].T[1][j])
        z.append(pfc[0].T[2][j])
    
ax.scatter(x,y,z,c=V,cmap=cm.jet,s=10)

#savefig("3D.svg")
```
